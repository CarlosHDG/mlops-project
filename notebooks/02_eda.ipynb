# EDA


```python
#imports
import pandas as pd
import numpy as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load cleaned data
df=pd.read_csv("../data/processed/cleaned_house_data_ds.csv")
print(df.shape)
print(df.head())
```

```python
#Statistics
print(df.describe())
```

```python
#Distributions
numerical_col=df.select_dtypes(include="number").columns
fig,axes=plt.subplots(nrows=2,ncols=3,figsize=(20,12))
axes_flatten=axes.flatten()
print(axes_flatten)

for i,col in enumerate(numerical_col):
    sns.histplot(data=df[col],kde=True,ax=axes_flatten[i])

plt.show
```

```python
#Correlation
plt.figure(figsize=(12,10))
corr_matrix=df.corr(numeric_only=True)

sns.heatmap(corr_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
#Scatter plot
fig,axes()
sns.scatterplot(x="bathrooms",y="price",data=df)
plt.title("Price vs Square Footage")
plt.show()
```

```python
#Bar chart
plt.figure(figsize=(5,4))
sns.countplot(x=df['location'])
plt.title("Number of houses by location")
plt.show()
```

```python
#Bar chart
plt.figure(figsize=(5,4))
sns.barplot(x="location",y="price",data=df)
plt.title("Average price by location")
plt.show()
```

```python
#Bar chart
plt.figure(figsize=(5,4))
sns.barplot(x="condition",y="price",data=df)
plt.title("Average price by condition")
plt.show()
```
